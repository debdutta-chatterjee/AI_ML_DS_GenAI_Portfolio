```python
# https://leetcode.com/problems/two-sum/

def twoSum(nums, target):
    counter = dict()
    for i,num in enumerate(nums):
        remaining = target-num
        if remaining in counter:
            return [counter[remaining], i]
        counter[num] = i
    return [-1,-1]

twoSum([2,7,11,15],9)
```

```python
# https://leetcode.com/problems/3sum/description/

def threeSum(nums):
    nums.sort()
    res =[]

    for i in range(len(nums)):
        if (i>0 and nums[i-1]==nums[i]):
            continue
        l,r = i+1, len(nums)-1

        while l<r:
            total = nums[i]+nums[l]+nums[r]
            if total>0:
                r-=1
            elif total<0:
                l+=1
            else:
                res.append([nums[i],nums[l],nums[r]])
                l+=1
                while (l<r and nums[l-1]==nums[l]):
                    l+=1
    return res

threeSum([-1,0,1,2,-1,-4])
```

```python
# https://leetcode.com/problems/3sum-closest/description/

def threeSumClosest(nums, target):
    nums.sort()
    sum_num = float('inf')

    for i in range(len(nums)):
        if i>0 and nums[i-1]==nums[i]:
            continue

        l,r = i+1,len(nums)-1

        while l<r:
            total_sum = nums[i]+nums[l]+nums[r]

            if abs(target-total_sum) < abs(target-sum_num):
                sum_num = total_sum

            if total_sum == target:
                return total_sum
            if total_sum >target:
                r-=1
            elif total_sum <target: 
                l+=1
    return sum_num

threeSumClosest([-1,2,1,-4],1)


```

```python
# https://leetcode.com/problems/remove-duplicates-from-sorted-array/description/

def removeDuplicates(nums):
    k=1
    for i in range(1,len(nums)):
        if not nums[i-1]==nums[i]:
            nums[k] = nums[i]
            k+=1
    return k

arr =[0,0,1,1,1,2,2,3,3,4]
k = removeDuplicates(arr)
print(arr[:k])
```

```python
# https://leetcode.com/problems/squares-of-a-sorted-array/description/

def sortedSquares(nums):
    n = len(nums)
    res = [0]*n
    l,r = 0 ,n-1
    pos = n-1

    while l<=r:
        if abs(nums[l]) > abs(nums[r]):
            res[pos] = nums[l]**2
            l+=1
        else:
            res[pos] = nums[r]**2
            r-=1
        pos-=1
    return res

sortedSquares([-4,-1,0,3,10])
```

```python
# https://leetcode.com/problems/sort-colors/description/

def sortColors(nums):
    l,r,m = 0,len(nums)-1,0

    while m<=r:
        if nums[m]==0:
            nums[l],nums[m] = nums[m],nums[l]
            l+=1
            m+=1
        elif nums[m]==1:
            m+=1
        elif nums[m]==2:
            nums[r],nums[m] = nums[m],nums[r]
            r-=1

arr = [2,0,2,1,1,0]
sortColors(arr)
print(arr)
```

```python
# https://leetcode.com/problems/largest-number/description/
from functools import cmp_to_key

def largestNumber(nums):

    for i, n in enumerate(nums):
        nums[i] = str(n)

    def compare(n1,n2):
        if n1+n2>n2+n1:
            return -1
        else:
            return 1
        
    nums = sorted(nums,key=cmp_to_key(compare))
    return str(int(''.join(nums)))


largestNumber([3,30,34,5,9])
```

```python
# https://leetcode.com/problems/wiggle-sort-ii/description/

def wiggleSort(nums):
    nums.sort()
    n = len(nums)
    result =[0]*n

    j = n-1

    for i in range(1,n,2):
        result[i] = nums[j]
        j-=1
    
    for i in range(0,n,2):
        result[i] = nums[j]
        j-=1

    for i in range(n):
        nums[i] = result[i]

    return nums

wiggleSort([3, 5, 2, 1, 6, 4])
```

```python
# https://leetcode.com/problems/merge-sorted-array/description/

def merge(nums1, m, nums2, n):
    p1 = m-1
    p2 = n-1
    k = len(nums1)-1

    while p1>=0 and p2>=0:
        if nums1[p1]>nums2[p2]:
            nums1[k] = nums1[p1]
            k-=1
            p1-=1
        else:
            nums1[k] = nums2[p2]
            k-=1
            p2-=1
    while p1>=0:
        nums1[k] = nums1[p1]
        k-=1
        p1-=1
    while p2>=0:
        nums1[k] = nums2[p2]
        k-=1
        p2-=1
    return nums1

merge([1,2,3,0,0,0],3,[2,5,6],3)
```

```python
# https://leetcode.com/problems/minimum-size-subarray-sum/description/

def minSubArrayLen(target, nums):
    left = 0
    min_length = float('inf')
    curr = 0

    for right in range(len(nums)):
        curr = curr+nums[right]

        while curr>=target:
            min_length = min(min_length,right-left+1)
            curr -= nums[left]
            left+=1
            
    return 0 if min_length ==float('inf') else min_length


minSubArrayLen(7,[2,3,1,2,4,3])
```

```python
# https://leetcode.com/problems/longest-mountain-in-array/description/

def longestMountain(arr):
    n = len(arr)
    max_length = 0
    i= 1

    while i<n-1:
        if arr[i-1]<arr[i]>arr[i+1]:
            left = i-1
            right = i+1

            while left>0 and arr[left-1]<arr[left]:
                left-=1
            while right<n-1 and arr[right]>arr[right+1]:
                right+=1
            max_length = max(max_length,right-left+1)
            i = right

        else:
            i+=1

    return max_length

longestMountain([2,1,4,7,3,2,5])
```

```python
# https://leetcode.com/problems/longest-continuous-increasing-subsequence/description/

def findLengthOfLCIS(nums):
    curr=1
    max_length = 1
    for i in range(1,len(nums)):
        if nums[i]>nums[i-1]:
            curr+=1
            max_length = max(max_length,curr)
        else:
            curr = 1

    return max_length

findLengthOfLCIS([1,3,5,4,7])
    
```

```python
# https://leetcode.com/problems/maximum-product-subarray/description/

def maxProduct(nums):
    result  = nums[0]
    max_so_far = nums[0]
    min_so_far  = nums[0]

    for i in range(1,len(nums)):
        curr = nums[i]
        temp_max = max(curr,curr*max_so_far,curr*min_so_far)
        temp_min = min(curr,curr*max_so_far,curr*min_so_far)

        max_so_far = temp_max
        min_so_far = temp_min

        result = max(max_so_far,result)

    return result


maxProduct([2,3,-2,4])
```
