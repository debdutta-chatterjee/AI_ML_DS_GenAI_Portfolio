```python
# https://leetcode.com/problems/find-the-index-of-the-first-occurrence-in-a-string/description/

def strStr(haystack, needle):
    i = 0
    j = i+len(needle)

    while j <= len(haystack):
        if haystack[i:j] == needle:
            return i
        else:
            i+=1
            j+=1
    return -1


strStr("sadbutsad","sad")
```

```python
# https://leetcode.com/problems/remove-duplicates-from-sorted-array/description/

def removeDuplicates(nums):
    k = 1

    for i in range(1,len(nums)):
        if not nums[i-1] == nums[i]:
            nums[k] = nums[i]
            k+=1
    return k

arr = [0,0,1,1,1,2,2,3,3,4]
k = removeDuplicates(arr)
print(arr[:k])
```

```python
# https://leetcode.com/problems/longest-substring-without-repeating-characters/description/

def lengthOfLongestSubstring(s):
    left = 0
    seen = set()
    max_len = 0

    for right in range(len(s)):
        curr = s[right]

        while curr in seen:
            seen.remove(s[left])
            left+=1

        seen.add(curr)
        max_len = max(max_len,right-left+1)

    return max_len

lengthOfLongestSubstring('abcabcbb')
```

```python
# https://leetcode.com/problems/find-all-anagrams-in-a-string/description/

def findAnagrams(s, p):

    res = []

    def char_to_index(c):
        return ord(c)-ord('a')

    s_map = [0]*26
    p_map = [0]*26

    for i in range(len(p)):
        p_map[char_to_index(p[i])] +=1
        s_map[char_to_index(s[i])] +=1

    if p_map == s_map:
        res.append(0)

    for i in range(len(p),len(s)):
        curr = s[i]
        s_map[char_to_index(curr)] +=1
        s_map[char_to_index(s[i-len(p)])] -=1

        if p_map == s_map:
            res.append(i-len(p)+1)
    return res
    

findAnagrams("cbaebabacd","abc")
```

```python
# https://leetcode.com/problems/group-anagrams/description/

def groupAnagrams(strs):

    freq_map = {}

    for str in strs:
        sorted_str = ''.join(sorted(str))
        if sorted_str in freq_map:
            freq_map[sorted_str].append(str)
        else:
            freq_map[sorted_str] = []
            freq_map[sorted_str].append(str)

    return list(freq_map.values())

groupAnagrams(["eat","tea","tan","ate","nat","bat"])
```

```python
# https://leetcode.com/problems/first-unique-character-in-a-string/description/

def firstUniqChar(s):
    freq_map = {}
    for c in s:
        freq_map[c] = freq_map.get(c,0)+1

    for i,c in enumerate(s):
        if freq_map.get(c,0) < 2:
            return i
    return -1

firstUniqChar('leetcode')
```

```python
# https://leetcode.com/problems/valid-anagram/description/

def isAnagram(s, t):
    s = ''.join(sorted(s))
    t = ''.join(sorted(t))

    return s == t

isAnagram("anagram","nagaram")
```

```python
# https://leetcode.com/problems/string-compression/description/

def compress(chars):
    read = 0
    write = 0

    while read < len(chars):
        char = chars[read]
        count = 0

        while read< len(chars) and chars[read] == char:
            read+=1
            count+=1

        chars[write] = char
        write+=1

        if count >1:
            for d in str(count):
                chars[write] = d
                write+=1
    return write

compress(["a","a","b","b","c","c","c"])
```
