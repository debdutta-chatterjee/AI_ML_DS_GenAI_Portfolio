```python
# 16. Reverse doubly linked list
class Node:
    def __init__(self,data):
        self.data = data
        self.prev = None
        self.next = None

def reverse_dll(head):
    curr =head
    while curr:
        next_node = curr.next
        prev_node = curr.prev

        curr.prev = next_node
        curr.next = prev_node

        tail = curr
        curr = next_node
        
    return tail

def printList(node):
    while node is not None:
        print(node.data, end=" ")
        node = node.next
    print()

if __name__ == "__main__":
  	
    # Create a doubly linked list:
    # 1 <-> 2 <-> 3 <-> 4
    head = Node(1)
    head.next = Node(2)
    head.next.prev = head
    head.next.next = Node(3)
    head.next.next.prev = head.next
    head.next.next.next = Node(4)
    head.next.next.next.prev = head.next.next

    print("Original Doubly Linked List")
    printList(head)
    head = reverse_dll(head)
    print("Reversed Doubly Linked List")
    printList(head)
```

```python
# 23. Add 2 numbers represented by linked lists
class Node:
    def __init__(self, val):
        self.data = val
        self.next = None

def reverse(head):
    prev = None
    while head:
        head.next, prev, head = prev, head, head.next
    return prev

def countNodes(head):
    count = 0
    while head:
        count += 1
        head = head.next
    return count

def addTwoLists(l1, l2):
    # Ensure l1 is longer
    if countNodes(l1) < countNodes(l2):
        l1, l2 = l2, l1

    # Reverse both lists
    l1 = reverse(l1)
    l2 = reverse(l2)

    dummy = Node(0)
    curr = dummy
    carry = 0

    while l1 or l2 or carry:
        sum_val = (l1.data if l1 else 0) + (l2.data if l2 else 0) + carry
        carry = sum_val // 10
        curr.next = Node(sum_val % 10)
        curr = curr.next

        if l1: l1 = l1.next
        if l2: l2 = l2.next

    return reverse(dummy.next)

def printList(head):
    while head:
        print(head.data, end=" ")
        head = head.next
    print()

# Example usage
if __name__ == "__main__":
    num1 = Node(1); num1.next = Node(2); num1.next.next = Node(3)
    num2 = Node(9); num2.next = Node(9); num2.next.next = Node(9)

    result = addTwoLists(num1, num2)
    printList(result)
```

```python
# 11. Insert in sorted linked list

# Node class
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

# Function to insert into sorted linked list
def insertSorted(head, value):
    
    curr = head

    while curr:
        if curr.data >= value:
            nodeInsert = Node(value)
            curr.data, nodeInsert.data = nodeInsert.data, curr.data
            nodeInsert.next = curr.next
            curr.next = nodeInsert
            break
        curr = curr.next
    return head

# ðŸ”§ Driver code
def printList(head):
    while head:
        print(head.data, end=" â†’ ")
        head = head.next
    print("None")

# Create initial sorted linked list: 10 â†’ 20 â†’ 30
head = Node(10)
head.next = Node(20)
head.next.next = Node(30)

print("Original list:")
printList(head)

# Insert new value
value_to_insert = 25
head = insertSorted(head, value_to_insert)

print(f"List after inserting {value_to_insert}:")
printList(head)
```
