```python
import numpy as np

docs = ['go india',
		'india india',
		'hip hip hurray',
		'jeetega bhai jeetega india jeetega',
		'bharat mata ki jai',
		'kohli kohli',
		'sachin sachin',
		'dhoni dhoni',
		'modi ji ki jai',
		'inquilab zindabad']
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer

tokenizer = Tokenizer(oov_token = '<unk>')
tokenizer.fit_on_texts(docs)
```

```python
tokenizer.word_counts
```

```python
sequence = tokenizer.texts_to_sequences(docs)
from keras.utils import pad_sequences

seq = pad_sequences(sequence,padding ='post')
seq
```

```python
len(tokenizer.word_index) + 1
```

```python
seq[0]
```

```python
from keras import Sequential
from keras.layers import Dense,SimpleRNN,Embedding
model1 = Sequential()
model1.add(Embedding(19,output_dim=2,input_length=5))
model1.build(input_shape=(None, 5))
model1.summary()
model1.compile('adam','accuracy')

single = seq[0]                  # shape (timesteps,)
single_input = np.expand_dims(single.astype('int32'), axis=0)  # shape (1, timesteps)

pred = model1.predict(single_input)
print(pred)
```

```python
from keras.datasets import imdb
from keras import Sequential
from keras.layers import Dense,SimpleRNN,Embedding,Flatten

(X_train,y_train), (X_test,y_test) = imdb.load_data()
```

```python
X_train = pad_sequences(X_train,padding='post',maxlen=50)
X_test = pad_sequences(X_test,padding='post',maxlen=50)
```

```python
model = Sequential()
model.add(SimpleRNN(32,input_shape =(50,1),return_sequences=False))
model.add(Dense(1,activation='sigmoid'))
model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics =['accuracy'])
model.fit(X_train,y_train,epochs =5,validation_data =(X_test,y_test))
```

```python
from keras.datasets import imdb
from tensorflow.keras.preprocessing.text import Tokenizer
from keras.utils import pad_sequences
from keras import Sequential
from keras.layers import Dense,SimpleRNN,Embedding,Flatten
```

```python
(X_train,y_train),(X_test,y_test) = imdb.load_data()
```

```python
X_train = pad_sequences(X_train,padding='post',maxlen=50)
X_test = pad_sequences(X_test,padding='post',maxlen=50)
X_train.shape
```

```python
vocab_size = int(max(X_train.max(), X_test.max())) + 1
print("vocab_size needed:", vocab_size)
```

```python
model = Sequential()
model.add(Embedding(88586, 2,input_length =50))
model.add(SimpleRNN(32,return_sequences=False))
model.add(Dense(1, activation='sigmoid'))
model.build(input_shape=(None, 50))
model.summary()
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['acc'])
history = model.fit(X_train, y_train,epochs=5,validation_data=(X_test,y_test))
```
