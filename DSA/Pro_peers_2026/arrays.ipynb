```python
# 167. Two Sum II - Input Array Is Sorted

def twoSum(numbers, target):
    i,j = 0, len(numbers)-1

    while i<=j:
        total_sum = numbers[i] + numbers[j]
        if total_sum == target:
            return [i,j]
        elif total_sum < target:
            i+=1
        else:
            j-=1
    return []


twoSum([2,7,11,15],9)
```

```python
# 3Sum

def threeSum(self, nums):
    nums.sort()
    res = []

    for i in range(len(nums)):
        if (i>0 and nums[i-1]==nums[i]):
            continue
        l,r = i+1,len(nums)-1

        while l<r:
            total = nums[i] + nums[l] + nums[r]

            if total >0:
                r-=1
            elif total <0:
                l+=1
            else:
                res.append([nums[i],nums[l],nums[r]])
                l+=1
                while (l<r and nums[l-1]==nums[l]):
                    l+=1
        return res
```

```python
# 3Sum Closest

def threeSumClosest(nums, target):
    nums.sort()
    closest = nums[0] + nums[1] + nums[2]

    for i in range(len(nums)-2):
        l,r = i+1,len(nums)-1

        while l<r:
            total = nums[l]+nums[i]+nums[r]

            if total == target:
                return total
            
            if abs(target - total) < abs(target - closest): 
                closest = total
            if total < target:
                l+=1
            else:
                r-=1
    return closest

threeSumClosest([-1,2,1,-4], 1)
```

```python
# Remove Duplicates from Sorted Array II

def removeDuplicates(nums):
    k = 2
    for i in range(2,len(nums)):        
        if nums[i] != nums[i-2]:
            nums[k] = nums[i]
            k+=1
    return k,nums

print(removeDuplicates([1,1,1,2,2,3]))
```

```python
# Sort Colors
def sortColors(nums):
    l,m,r = 0,0,len(nums)-1

    while m<=r:
        if nums[m] == 0:
            nums[l],nums[m] = nums[m], nums[l]
            l+=1
            m+=1

        elif nums[m] == 1:
            m+=1
        
        elif nums[m] == 2:
            nums[r],nums[m] = nums[m], nums[r]
            r-=1
    return nums

sortColors([2,0,2,1,1,0])
```

```python
#  Squares of a Sorted Array

def sortedSquares(nums):
    n = len(nums)-1
    i,j = 0,len(nums)-1
    res = [0]*len(nums)
    while i<=j:
        if abs(nums[i]) > abs(nums[j]):
            res[n] = nums[i]*nums[i]
            i+=1
        else:
            res[n] = nums[j]*nums[j]
            j-=1
        n-=1
    return res


sortedSquares([-4,-1,0,3,10])
```

```python
# Merge Sorted Array

def merge(nums1, m, nums2, n):
    pos = m+n-1

    i,j = m-1,n-1

    while i>=0 and j>=0:
        if nums1[i]>nums2[j]:
            nums1[pos] = nums1[i]
            i-=1
        else:
            nums1[pos] = nums2[j]
            j-=1
        pos-=1
    
    while i>=0:
        nums1[pos] = nums1[i]
        pos-=1
        i-=1

    while j>=0:
        nums1[pos] = nums2[j]
        pos-=1
        j-=1     

    return nums1

merge([1,2,3,0,0,0],3,[2,5,6],3)
```

```python
# Longest Mountain in Array

def longestMountain(arr):
    if len(arr)<3:
        return 
    
    n = len(arr)
    max_length = 0
    i = 1

    while i < n-1:
        if arr[i-1]<arr[i]>arr[i+1]:
            l = i-1
            r = i+1

            while l >0 and arr[l-1]<arr[l]:
                l-=1
            while r < n-1 and arr[r]>arr[r+1]:
                r+=1
            
            max_length = max(max_length,r-l+1)
            i= r
        else:
            i+=1
    return max_length

longestMountain([2,1,4,7,3,2,5])
```

```python
# Minimum Absolute Difference

def minimumAbsDifference(arr):
    arr.sort()

    min_dis = float('inf') 
    for i in range(1, len(arr)): 
        min_dis = min(min_dis, arr[i] - arr[i-1])
    res = []
    for i in range(1,len(arr)):
        if arr[i] -arr[i-1] == min_dis:
            res.append([arr[i-1],arr[i]])
    return res

minimumAbsDifference([4,2,1,3])
```

```python
# Maximum Product of Three Numbers

def maximumProduct(nums):
    if len(nums) <3:
        return 
    nums.sort(reverse=True)

    return max(nums[2] * nums[1] * nums[0],nums[-1]*nums[-2]*nums[0])

maximumProduct([1,2,3,4])
```

```python
# 561. Array Partition

def arrayPairSum(nums):
    nums.sort() 
    return sum(nums[::2])

arrayPairSum([6,2,6,5,1,2])
```

```python
# 581. Shortest Unsorted Continuous Subarray

def findUnsortedSubarray(nums):
    sorted_nums = sorted(nums)

    start, end = len(nums),0
    for i in range(0,len(nums)):
        if nums[i]!= sorted_nums[i]:
            start = min(start,i)
            end = max(end,i)

    return end-start+1 if end-start >0 else 0

findUnsortedSubarray([2,6,4,8,10,9,15])
```
