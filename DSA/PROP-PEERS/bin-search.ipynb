```python
# https://leetcode.com/problems/search-insert-position/description/

def searchInsert(nums, target):
    l,r = 0 ,len(nums)-1

    while l<=r:
        m = (l+r)//2

        if nums[m] == target:
            return m
        elif nums[m] > target:
            r=m-1
        else:
            l = m+1
    return l


searchInsert([1,3,5,6],5)
```

```python
# https://leetcode.com/problems/find-minimum-in-rotated-sorted-array/description/

def findMin(nums):
    l,r = 0,len(nums)-1

    while l<r:
        m = (l+r)//2
        if nums[m]>nums[r]:
            l = m+1
        else:
            r = m
    return nums[l]

findMin([3,4,5,1,2])
```

```python
# https://leetcode.com/problems/peak-index-in-a-mountain-array/description/

def peakIndexInMountainArray(arr):
    l,r = 0,len(arr)-1

    while l<r:
        m = (l+r)//2
        if arr[m]<arr[m+1]:
            l = m+1
        else:
            r = m
    return l


peakIndexInMountainArray([0,10,5,2])
```

```python
# https://leetcode.com/problems/find-smallest-letter-greater-than-target/description/

def nextGreatestLetter(letters, target):
    l,r = 0, len(letters)-1

    while l<r:
        m =(l+r)//2
        if letters[m]  <=target:
            l = m+1
        else:
            r = m
    return letters[l] if l<len(letters) else letters[0]

nextGreatestLetter(["c","f","j"],"z")
```

```python
# https://leetcode.com/problems/valid-perfect-square/description/

def isPerfectSquare(num):
    l,r = 2, num//2

    while l<=r:
        m = (l+r)//2

        if m*m == num:
            return True
        elif m*m > num:
            r = m-1
        else:
            l = m+1
    return False


isPerfectSquare(256)
```

```python
# https://leetcode.com/problems/kth-smallest-element-in-a-sorted-matrix/

def kthSmallest(matrix, k):
    m = len(matrix)
    n = len(matrix[0])

    def count_less_than(m):
        count = 0
        r,c = len(matrix)-1,0
        while r>=0 and c <len(matrix):
            if matrix[r][c]<m:
                count+=r+1
                c+=1
            else:
                r-=1
        return count

    l,h = matrix[0][0],matrix[-1][-1]
    while l<h:
        m = (l+h)//2

        if count_less_than(m) <k:
            l = m+1
        else:
            h = m
    return l
```
