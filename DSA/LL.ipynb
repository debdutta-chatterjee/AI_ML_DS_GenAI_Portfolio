```python
# 6. Reverse linked list
 
def reverseList(head):
    prev = None
    curr = head

    while curr:
        next_node = curr.next
        curr.next = prev
        prev = curr
        curr =next_node
    head = prev
    return head

def ll_print(head):
    curr =head
    while curr:
        print(curr.value,end=' ')
        curr = curr.next
    print()


if __name__ == '__main__':
# Create a hard-coded linked list:
# 1 -> 2 -> 3 -> 4 -> 5
    head = Node(1)
    head.next = Node(2)
    head.next.next = Node(3)
    head.next.next.next = Node(4)
    head.next.next.next.next = Node(5)
    ll_print(head)
    head = reverseList(head)

    ll_print(head)

```

```python
#7. Detect loop in a linked list- McKinsey

def detectLoop(head):
    unique = set()
    length = 0

    curr = head
    while curr:
        unique.add(curr)
        curr = curr.next
        length+=1
        if len(unique) != length:
            return True

    return len(unique) != length


def detectLoop_optimum(head):
    fast =slow =head

    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next

        if slow == fast:
            return True
    return False



if __name__ == "__main__":

    # Create a hard-coded linked list:
    # 1 -> 3 -> 4
    head = Node(1)
    head.next = Node(3)
    head.next.next = Node(4)
  
    # Create a loop
    head.next.next.next = head.next

    if detectLoop(head):
        print("true")
    else:
        print("false")
```
