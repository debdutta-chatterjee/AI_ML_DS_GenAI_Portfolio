```python
# https://leetcode.com/problems/middle-of-the-linked-list/description/

class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution(object):
    def middleNode(self, head):
        fp = sp = head
        
        while fp and fp.next:
            sp = sp.next
            fp = fp.next.next
        return sp
    
# ---------------- Runner ----------------
def build_linked_list(values):
    """Helper to build a linked list from a Python list."""
    dummy = ListNode(0)
    current = dummy
    for v in values:
        current.next = ListNode(v)
        current = current.next
    return dummy.next

def print_linked_list(head):
    """Helper to print linked list values."""
    current = head
    while current:
        print(current.val, end=" -> " if current.next else "")
        current = current.next
    print()

# Example usage:
if __name__ == "__main__":
    # Build a linked list [1, 2, 3, 4, 5, 6]
    head = build_linked_list([1, 2, 3, 4, 5, 6])
    print("Original Linked List:")
    print_linked_list(head)

    # Find middle node
    mid = Solution().middleNode(head)
    print("\nMiddle Node Value:", mid.val)

    # Print from middle to end
    print("Linked List starting from Middle:")
    print_linked_list(mid)
```

```python
# https://leetcode.com/problems/intersection-of-two-linked-lists/description/

class ListNode(object):
    def __init__(self, x):
        self.val = x
        self.next = None

class Solution(object):
    def getIntersectionNode(self, headA, headB):
        if not headA or not headB:
            return None
        
        pA = headA
        pB = headB

        while pA != pB:
            pA = pA.next if pA else headB
            pB = pB.next if pB else headA

        return pA

# ---------------- Runner ----------------
def print_linked_list(head):
    """Helper to print linked list values."""
    current = head
    while current:
        print(current.val, end=" -> " if current.next else "")
        current = current.next
    print()

if __name__ == "__main__":
    # Create shared intersection part: c1 -> c2 -> c3
    c1 = ListNode(8)
    c2 = ListNode(9)
    c3 = ListNode(10)
    c1.next = c2
    c2.next = c3

    # Create list A: a1 -> a2 -> c1 -> c2 -> c3
    a1 = ListNode(1)
    a2 = ListNode(2)
    a1.next = a2
    a2.next = c1
    headA = a1

    # Create list B: b1 -> b2 -> b3 -> c1 -> c2 -> c3
    b1 = ListNode(3)
    b2 = ListNode(4)
    b3 = ListNode(5)
    b1.next = b2
    b2.next = b3
    b3.next = c1
    headB = b1

    # Print both lists
    print("List A:")
    print_linked_list(headA)
    print("List B:")
    print_linked_list(headB)

    # Find intersection
    intersection = Solution().getIntersectionNode(headA, headB)
    if intersection:
        print("\nIntersection at node with value:", intersection.val)
    else:
        print("\nNo intersection found.")
```

```python
# https://leetcode.com/problems/linked-list-cycle/description/

class ListNode(object):
    def __init__(self, x):
        self.val = x
        self.next = None

class Solution():
    def hasCycle(self, head):
        """
        :type head: ListNode
        :rtype: bool
        """
        sp, fp = head,head

        while fp and fp.next:
            sp = sp.next
            fp = fp.next.next

            if sp == fp:
                return True
            
        
        return False

# Helper to build a cycle list
def build_cycle_list():
    head = ListNode(1)
    node2 = ListNode(2)
    node3 = ListNode(3)
    node4 = ListNode(4)
    
    head.next = node2
    node2.next = node3
    node3.next = node4
    node4.next = node2  # cycle back to node2
    
    return head

# Test
head = build_cycle_list()
print(Solution().hasCycle(head))  # True

# Non-cycle list
head2 = ListNode(1)
head2.next = ListNode(2)
print(Solution().hasCycle(head2))  # False
```

```python
# https://leetcode.com/problems/remove-nth-node-from-end-of-list/description/

class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution(object):
    def removeNthFromEnd(self, head, n):
        dummy = ListNode(0, head)   # Add a dummy node before head
        sp, fp = dummy, dummy

        for i in range(n+1):
            fp = fp.next

        while fp:
            sp = sp.next
            fp = fp.next

        sp.next = sp.next.next

        return dummy.next
    
# Helper function to create a linked list from a list
def create_linked_list(values):
    dummy = ListNode(0)
    current = dummy
    for val in values:
        current.next = ListNode(val)
        current = current.next
    return dummy.next

# Helper function to print linked list
def print_linked_list(head):
    current = head
    result = []
    while current:
        result.append(str(current.val))
        current = current.next
    print(" -> ".join(result))

# Runner code
if __name__ == "__main__":
    # Example: Linked list = [1, 2, 3, 4, 5], remove 2nd node from end
    values = [1, 2, 3, 4, 5]
    head = create_linked_list(values)
    print("Original list:")
    print_linked_list(head)

    n = 2
    head = Solution().removeNthFromEnd(head, n)
    print(f"\nList after removing {n}th node from end:")
    print_linked_list(head)
```

```python
# https://leetcode.com/problems/odd-even-linked-list/description/

class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution(object):
    def oddEvenList(self, head):
        odd, even = head, head.next
        even_head = even

        while even:

            odd.next = even.next
            even.next = even.next.next

            odd = odd.next
            even = even.next

        odd.next = even_head

        return head
    
# Helper function to create a linked list from a list
def create_linked_list(values):
    dummy = ListNode(0)
    current = dummy
    for val in values:
        current.next = ListNode(val)
        current = current.next
    return dummy.next

# Helper function to print linked list
def print_linked_list(head):
    current = head
    result = []
    while current:
        result.append(str(current.val))
        current = current.next
    print(" -> ".join(result))

# Runner code
if __name__ == "__main__":
    # Example: Linked list = [1, 2, 3, 4, 5]
    values = [1, 2, 3, 4, 5]
    head = create_linked_list(values)
    print("Original list:")
    print_linked_list(head)

    head = Solution().oddEvenList(head)
    print("\nList after grouping odd and even indices:")
    print_linked_list(head)
```

```python
# https://leetcode.com/problems/linked-list-cycle-ii/description/

class ListNode(object):
    def __init__(self, x):
        self.val = x
        self.next = None

class Solution(object):
    def detectCycle(self, head):

        if not head or not head.next:
            return None

        sp, fp, p = head,head,head

        while fp and fp.next:
            sp = sp.next
            fp = fp.next.next

            if sp == fp:
                break

        if sp!=fp:
            return None
        
        while p!=sp:
            p = p.next
            sp = sp.next

        return sp
```

```python
# https://leetcode.com/problems/merge-two-sorted-lists/description/

class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution():
    def mergeTwoLists(self,list1, list2):
        head_l1, head_l2 = list1,list2
        dummy = ListNode(0)
        curr = dummy

        while head_l1 and head_l2:
            if head_l1.val <= head_l2.val:
                curr.next = head_l1
                head_l1 = head_l1.next
            else:
                curr.next = head_l2
                head_l2 = head_l2.next
            curr = curr.next

        if not head_l1 and not head_l2:
            return dummy.next
        
        rem = head_l1 if head_l1 else head_l2

        while rem:
            curr.next = rem
            curr = curr.next
            rem = rem.next
        return dummy.next

# Runner code
if __name__ == "__main__":
    # Helper function to create linked list from Python list
    def create_linked_list(arr):
        dummy = ListNode(0)
        curr = dummy
        for val in arr:
            curr.next = ListNode(val)
            curr = curr.next
        return dummy.next

    # Helper function to print linked list
    def print_linked_list(head):
        while head:
            print(head.val, end=" -> ")
            head = head.next
        print("None")

    # Example input lists
    list1 = create_linked_list([1, 2, 4])
    list2 = create_linked_list([1, 3, 4])

    # Call merge function (assumed implemented)
    merged_head = Solution().mergeTwoLists(list1, list2)

    # Print result
    print_linked_list(merged_head)
```

```python
# https://leetcode.com/problems/add-two-numbers/description/

class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution(object):
    def addTwoNumbers(self, l1, l2):
        c =0
        dummy = ListNode(0)
        ptr = dummy

        while c!=0 or l1 or l2:
            val1 = l1.val if l1 else 0
            val2 = l2.val if l2 else 0

            c += val1 + val2
            ptr.next = ListNode(c%10)
            c = c//10
            if l1: 
                l1 = l1.next 
            if l2: 
                l2 = l2.next
            ptr = ptr.next

        return dummy.next   

# Runner code
if __name__ == "__main__":
    # Helper function to create linked list from Python list
    def create_linked_list(arr):
        dummy = ListNode(0)
        curr = dummy
        for val in arr:
            curr.next = ListNode(val)
            curr = curr.next
        return dummy.next

    # Helper function to print linked list
    def print_linked_list(head):
        while head:
            print(head.val, end=" -> ")
            head = head.next
        print("None")

    # Example input lists (representing 342 and 465)
    l1 = create_linked_list([2, 4, 3])  # 342
    l2 = create_linked_list([5, 6, 4])  # 465
    
    # Run addTwoNumbers
    sol = Solution()
    result = sol.addTwoNumbers(l1, l2)

    # Print result (should represent 807 → [7, 0, 8])
    print_linked_list(result)
```

```python
# https://leetcode.com/problems/partition-list/description/

class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution(object):
    def partition(self, head, x):
        """
        :type head: Optional[ListNode]
        :type x: int
        :rtype: Optional[ListNode]
        """
        prev= ListNode(0)
        ptr1 = prev
        nxt = ListNode(0)
        ptr2 = nxt
        ptr = head

        while ptr:
            if ptr.val<x:
                ptr1.next = ptr
                ptr1 = ptr1.next
            else:
                ptr2.next = ptr
                ptr2 = ptr2.next
            ptr = ptr.next
        if ptr2:
            ptr2.next = None
        ptr1.next = nxt.next
        return prev.next
    

# Runner code
if __name__ == "__main__":
    # Helper function to create linked list from Python list
    def create_linked_list(arr):
        dummy = ListNode(0)
        curr = dummy
        for val in arr:
            curr.next = ListNode(val)
            curr = curr.next
        return dummy.next

    # Helper function to print linked list
    def print_linked_list(head):
        while head:
            print(head.val, end=" -> ")
            head = head.next
        print("None")
    
    l1 = create_linked_list([1,4,3,2,5,2])
    
    # Run addTwoNumbers
    sol = Solution()
    result = sol.partition(l1,3)

    # Print result (should represent 807 → [7, 0, 8])
    print_linked_list(result)
```

```python
# https://leetcode.com/problems/remove-duplicates-from-sorted-list/description/

class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution():
    def deleteDuplicates(self, head):

        seen = []
        ptr = head
        dummy =  ListNode(0)
        curr = dummy

        while ptr:
            if not ptr.val in seen:
                curr.next = ptr
                curr = curr.next  
                seen.append(ptr.val)          
            ptr = ptr.next
        curr.next = None

        return dummy.next
    
l1 = create_linked_list([1,4,3,2,5,2])

sol = Solution()
result = sol.deleteDuplicates(l1)

print_linked_list(result)
```
